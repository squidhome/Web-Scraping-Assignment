```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://books.toscrape.com/catalogue/category/books/romance_8/index.html'
```

```python
index_html = requests.get(url).text
```

```python
index_html
```

```python
index = BeautifulSoup(index_html, 'html.parser')
```

```python
index
```

```python
books_item = index.find_all('li', class_ = 'col-xs-6 col-sm-4 col-md-3 col-lg-3')
```

```python
books_item
```

```python
book_html = [x.find('h3') for x in books_item]
```

```python
book_html
```

```python
book_title = [x.find('a').get('title') for x in book_html]
```

```python
book_title
```

```python
book_links = ['http://books.toscrape.com/catalogue/' + x.find('a').get('href').lstrip('../') for x in books_item]
```

```python
book_links
```

```python
book_rating = [x.find('p').get('class')[1] for x in books_item]
```

```python
book_rating
```

```python
book_price = ['£' + x.find('p', class_ = 'price_color').getText().lstrip('Â£') for x in books_item]
```

```python
book_price
```

```python
import pandas as pd
books_df = pd.DataFrame({'Title': book_title,
                         'Link': book_links,
                         'Price': book_price,
                         'Rating': book_rating
                        })
```
